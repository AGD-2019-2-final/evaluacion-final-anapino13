```python
%%pig
fs -rm output/*
fs -rmdir  output
fs -rm -f -r data.tsv
fs -put data.tsv

u = LOAD 'data.tsv'
    AS (f1:CHARARRAY, f2:CHARARRAY, f3:INT);
orden = ORDER u BY f1,f3;
DUMP orden;


STORE orden INTO 'output';

fs -copyToLocal output
```
